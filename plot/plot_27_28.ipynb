```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1024 = pd.read_csv('query28/1024_sf1.csv')
df2048 = pd.read_csv('query28/2048_sf1.csv')
df4096 = pd.read_csv('query28/4096_sf1.csv')
plt.plot(df1024['N'], df1024['runtime'], label='Hash Table Size = 1024')
plt.plot(df2048['N'], df2048['runtime'], label='Hash Table Size = 2048')
plt.plot(df4096['N'], df4096['runtime'], label='Hash Table Size = 4096')
plt.legend()
plt.xlabel("N - the number of probings")
plt.ylabel("Query Runtime (ms)")
plt.title("The Query28-SF1 Runtime with different N (hash table probing times)")
plt.rcParams['figure.figsize'] = [50, 8]
plt.show()
```
